```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D 
plt.rcParams.update({'font.size': 12})

# constants
D = 0.1 # diffusivity [m^2 / s]
L = 1 # length [m]
T = 1 # time [s]

t_space_pos_num = 2500
x_space_pos_num = 30

# functions
def solve(n: int, tsteps: int, dx: np.float64, dt: np.float64) -> list:
    '''Solves heat equation using Euler method.'''
    r = D * dt / dx ** 2
    u = [0 for _ in range(n)]
    res = []
    
    # boundary conditions
    u[0] = u[n - 1] = 0
    
    # initial condition
    for i in range(1, n):
        x = i * dx
        u[i] = np.sin(np.pi * x)
    
    for t in range(1, tsteps + 1):
        unew = [0 for _ in range(n)]
        for i in range(1, n - 1):
            unew[i] = r * u[i - 1] + (1 - 2 * r) * u[i] + r * u[i + 1]
        res.append(unew)
        u = unew
    return res

def plot(x: np.ndarray, t: np.ndarray) -> None:
    X, T = np.meshgrid(x, t)
    fig = plt.figure(figsize=(15,7))
    ax = fig.add_subplot(1, 2, 1)
    ax.pcolormesh(x, t, u, cmap='coolwarm')
    ax.set_xlabel('x')
    ax.set_ylabel('t')
    
    ax = fig.add_subplot(122, projection='3d')
    ax.plot_surface(X, T, np.array(u), cmap='coolwarm')
    ax.set_xlabel('x')
    ax.set_ylabel('t')
    ax.set_zlabel('u(x,t)')

    plt.show()
```

```python
x, dx = np.linspace(0, L, x_space_pos_num, retstep=True)
t, dt = np.linspace(0, T, t_space_pos_num, retstep=True)
u = solve(x_space_pos_num, t_space_pos_num, dx, dt)
```

```python
plot(x, t)
```
