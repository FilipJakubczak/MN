```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def approx_polynomial(x: np.ndarray, y: np.ndarray, degree: int = 3) -> np.ndarray:
    xsum = np.zeros(2 * degree + 1)
    A = np.zeros((degree + 1, degree + 1))
    b = np.zeros(degree + 1)
    for j in range(0, 2 * degree + 1):
        xsum[j] = np.sum(x ** j)
    for j in range(0, degree + 1):
        for i in range(0, degree + 1):
            A[i][j] = xsum[i + j]
        b[j] = np.sum(x ** j * y)
    return np.linalg.solve(A, b)
```

```python
points_x = np.array([1, 2, 3, 4, 5])
points_y = np.array([52, 5, -5, -40, 10])
```

```python
fig, ax = plt.subplots(figsize=(15, 10))
ax.scatter(points_x, points_y, label='Set of points', color='blue')
x = np.arange(points_x[0] - 1, points_x[-1] + 1, 0.01)
coef = approx_polynomial(points_x, points_y, degree=3)
y = [np.sum([value * xj ** (len(coef) - 1 - i) for i, value in enumerate(coef[::-1])]) for xj in x]
ax.plot(x, y, label='Approximated polynomial', color='red')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
