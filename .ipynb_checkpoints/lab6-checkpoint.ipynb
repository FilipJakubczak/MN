```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
def dx(x:float, y: float, z: float) -> float:
    return 10 * (y - x)

def dy(x:float, y: float, z: float) -> float:
    return x * (28 - z) - y

def dz(x: float, y: float, z: float) -> float:
    return x * y - 8 / 3 * z

def diff_k(k: list, var: str, step: float) -> float:
    if var == 'x':
        k = k[0]
    elif var == 'y':
        k = k[1]
    elif var == 'z':
        k = k[2]
    else:
        return None
    return 1 / 2 * k * step

def diff_var(k: list, var: str, step: float) -> float:
    if var == 'x':
        i = 0
    elif var == 'y':
        i = 1
    elif var == 'z':
        i = 2
    else:
        return None
    return 1 / 6 * (k[0,i] + 2 * k[1,i] + 2 * k[2,i] + k[3,i]) * time[1]

def solve(f: np.ndarray, x0: float, y0: float, z0: float,
          time: tuple) -> dict:
    '''Solves the given system of differential equations using Runge-Kutta 4th order method.'''
    x, y, z = x0, y0, z0
    trajectory = {'x': [], 'y': [], 'z': []}
    k = np.zeros((4, 3))
    for t in time[0]:
        kwargs = [{'x': x, 'y': y, 'z': z},
                  {'x': x + diff_k(k[0], 'x', time[1]), 
                   'y': y + diff_k(k[0], 'y', time[1]), 
                   'z': z + diff_k(k[0], 'z', time[1])},
                  {'x': x + diff_k(k[1], 'x', time[1]),
                   'y': y + diff_k(k[1], 'y', time[1]),
                   'z': z + diff_k(k[1], 'z', time[1])},
                  {'x': x + diff_k(k[2], 'x', time[1]),
                   'y': y + diff_k(k[2], 'y', time[1]),
                   'z': z + diff_k(k[2], 'z', time[1])}]
        for i in range(0, 4):
            k[i] = [dx(**kwargs[i]), dy(**kwargs[i]), dz(**kwargs[i])]
        x += diff_var(k, 'x', time[1])
        y += diff_var(k, 'y', time[1])
        z += diff_var(k, 'z', time[1])
        trajectory['x'].append(x)
        trajectory['y'].append(y)
        trajectory['z'].append(z)
    return trajectory

def plot(points: dict, time: tuple) -> None:
    '''Plots system's trajectory.'''
    fig = plt.figure(figsize=(15, 10))
    ax = fig.add_subplot(111, projection='3d')
    ax.plot(trajectory['x'], trajectory['y'], trajectory['z'])
    ax.set_xlim(-20, 20)
    ax.set_ylim(-30, 30)
    ax.set_zlim(0, 50)
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_zlabel('z')
    ax.view_init(30, 235)
    plt.title('Result for t ∈ ⟨{:.0f},{:.0f}⟩ and {:.0f} time units.'.format(
    time[0].min(), time[0].max(), (time[0].max() - time[0].min()) / time[1] + 1))
    plt.show()
```

```python
# Initial conditions
x0 = 1
y0 = 1
z0 = 1
# Intervals
time = np.linspace(0, 100, 10000, retstep=True)
# Functions
fun = np.array([dx, dy, dz])
```

```python
trajectory = solve(fun, x0, y0, z0, time)
plot(trajectory, time)
```
